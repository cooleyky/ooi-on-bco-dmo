```python
# Import modules
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Load data from notebook 1 into a Pandas Dataframe
df = pd.read_csv("../data/external/irminger_sea_subset.csv")
df.head() # Print the first 5 rows of the DataFrame
```

```python
# Query cast data for a particular target mooring and reset the index
mooring = "SUMO"
sumo = df.query("`Target_Asset`==@mooring")
sumo.reset_index(inplace=True)
sumo.head()
```

```python
print(sumo.columns)
```

```python
some_vars = sumo[["CTD_Depth", "Discrete_Oxygen", "Discrete_Chlorophyll", "Discrete_Nitrate"]]
```

```python
some_vars
```

```python
cast_cmap = mpl.colormaps['Accent'].reversed() # Reversed in attempt to improve colorblind-friendliness
```

```python
time = pd.to_datetime(sumo["CTD_Bottle_Closure_Time"], yearfirst=True, utc=True).values
```

```python
# Create the subplots
fig, axs = plt.subplots(3, 1, sharex=True, layout="tight")
# Discrete Oxygen
plt.subplot(3,1,1)
plt.scatter(x=time, y=sumo["Discrete_Oxygen"], c=sumo["Cast"].astype(float), cmap=cast_cmap, edgecolors="none")
plt.grid()
axs[0].set_ylabel("Discrete Oxygen")
# CTD Fluorescence, I wanted to use Discrete Chlorophyll but one of the values couldn't be parsed as a float ("05/01")
plt.subplot(3,1,2)
plt.scatter(x=time, y=sumo["CTD_Fluorescence"], c=sumo["Cast"].astype(float), cmap=cast_cmap, edgecolors="none")
plt.grid()
axs[1].set_ylabel("CTD Fluorescence")
# Discrete Nitrate
plt.subplot(3,1,3)
plt.scatter(x=time, y=sumo["Discrete_Nitrate"], c=sumo["Cast"].astype(float), cmap=cast_cmap, edgecolors="none")
plt.grid()
axs[2].set_ylabel("Discrete Nitrate")
axs[2].set_xlabel("Time")
```
